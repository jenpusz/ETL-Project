## Query to join child poverty and child mortality tables

```python
pd.read_sql_query("""select M.country_code_name ,M.observation as mortality_observation ,
M.unit_of_measure as mortality_unit_of_measure,
P.observation as poverty_observation , P.unit_of_measure as poverty_unit_of_measure,
M.timeperiod,
M.sex
from child_mortality M
Join child_poverty P
on M.country_code_name = P.country_code_name
and M.timeperiod = p.timeperiod
and M.sex=p.sex""", con=engine).head()
```
