### Store CSV into DataFrame

```python
csv_file_child_mortality = "Resources/UNICEF_child_mortality.csv"

UNICEF_child_mortality_data_df = pd.read_csv(csv_file_child_mortality)
UNICEF_child_mortality_data_df.head(5)
```

### Create new data with select columns

```python
new_UNICEF_child_mortality_data_df = UNICEF_child_mortality_data_df[["REF_AREA:Country","INDICATOR:Indicator","SEX:Sex","TIME_PERIOD:Time period","OBS_VALUE:Observation Value","UNIT_MEASURE:Unit of measure"]]
new_UNICEF_child_mortality_data_df.head(5)
```

```python
new_UNICEF_child_mortality_data_df = new_UNICEF_child_mortality_data_df.rename(columns={'REF_AREA:Country':'country_code_name','INDICATOR:Indicator':'indicator',
'OBS_VALUE:Observation Value':'observation','UNIT_MEASURE:Unit of measure':'unit_of_measure','SEX:Sex':'sex','TIME_PERIOD:Time period':'timeperiod'},inplace = False)
```
